```python
from deepface import DeepFace
import os
import json

dataset_path = "D:\kandupidtham\ArjunProjectDefense\images"
embeddings = {}

for person_folder in os.listdir(dataset_path):
    person_path = os.path.join(dataset_path, person_folder)
    print('pers' + person_path)
    if os.path.isdir(person_path):
        embeddings[person_folder] = []
        for image_file in os.listdir(person_path):
            image_path = os.path.join(person_path, image_file)
            embedding = DeepFace.represent(image_path, model_name="Facenet")
            embeddings[person_folder].append(embedding)

print("embedding -------------" + json.dump(embeddings, f))
with open("embeddings.json", "w") as f:
    json.dump(embeddings, f)
```

```python
from deepface.commons import distance
import json

# Load saved embeddings
with open("embeddings.json", "r") as f:
    embeddings = json.load(f)

uploaded_image_path = "D:\kandupidtham\ArjunProjectDefense\arju_test.jpg"
uploaded_embedding = DeepFace.represent(uploaded_image_path, model_name="Facenet")

threshold = 0.4  # Set based on testing
best_match = None
best_score = float("inf")

for person, person_embeddings in embeddings.items():
    for embedding in person_embeddings:
        score = distance.findCosineDistance(uploaded_embedding, embedding)
        if score < best_score and score < threshold:
            best_match = person
            best_score = score

if best_match:
    print(f"Match found: {best_match}")
else:
    print("No match found.")
```
